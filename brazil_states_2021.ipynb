```python
# selecionando colunas de interesse e guardando em um novo DataFrame

clean_df = df.iloc[:,17:19]
```

```python
clean_df.head()
```

#### Ajustando nome das colunas

```python
# ajustando os nomes das colunas 

nomes = clean_df.columns.to_list()
nomes
```

```python
nomes_ajustados = []

for nome in nomes:
    
    nome_ajustado = (nome.split(sep=',')[-1].strip().replace("'", "").replace(")", "").lower())
  
    nomes_ajustados.append(nome_ajustado)
    
# O método split() divide uma string em uma lista.
```

```python
# verificando

nomes_ajustados
```

```python
clean_df.columns = nomes_ajustados
clean_df.head()
```

```python
# agrupando por nível e faixa salarial

clean_df.groupby(['nivel','faixa salarial'])['quantidade'].sum().reset_index()
```

```python
faixa_por_nivel = clean_df.groupby(['nivel','faixa salarial'], as_index=False)['quantidade'].sum()
faixa_por_nivel.head()
```

```python
# pd.DataFrame aceita como argumento um dicionário

faixa_por_nivel['faixa salarial'].drop_duplicates().to_dict()
```

```python
faixa_salarial_ordenada = [
  {'ordem': 0,'faixa salarial': 'Menos de R$ 1.000/mês'},
  {'ordem': 1,'faixa salarial': 'de R$ 1.001/mês a R$ 2.000/mês'},
  {'ordem': 2,'faixa salarial': 'de R$ 2.001/mês a R$ 3000/mês'},
  {'ordem': 3,'faixa salarial': 'de R$ 3.001/mês a R$ 4.000/mês'},
  {'ordem': 4,'faixa salarial': 'de R$ 4.001/mês a R$ 6.000/mês'},
  {'ordem': 5,'faixa salarial': 'de R$ 6.001/mês a R$ 8.000/mês'},
  {'ordem': 6,'faixa salarial': 'de R$ 8.001/mês a R$ 12.000/mês'},
  {'ordem': 7,'faixa salarial': 'de R$ 12.001/mês a R$ 16.000/mês'},
  {'ordem': 8,'faixa salarial': 'de R$ 16.001/mês a R$ 20.000/mês'},
  {'ordem': 9,'faixa salarial': 'de R$ 20.001/mês a R$ 25.000/mês'},
  {'ordem': 10,'faixa salarial': 'de R$ 25.001/mês a R$ 30.000/mês'},
  {'ordem': 11,'faixa salarial': 'de R$ 30.001/mês a R$ 40.000/mês'},
  {'ordem': 12,'faixa salarial': 'Acima de R$ 40.001/mês'}
]
```

```python
faixa_salarial_ordenada_df = pd.DataFrame(faixa_salarial_ordenada)
```

```python
faixa_salarial_ordenada_df
```

```python
# join para mesclar coluna de ordenação ao dataframe original

faixa_por_nivel = faixa_por_nivel.merge(faixa_salarial_ordenada_df, on=['faixa salarial'], how='left')
faixa_por_nivel.head()
```

```python
# ordenando por nível de ordem

faixa_por_nivel = faixa_por_nivel.sort_values(['nivel', 'ordem'], ascending=[True, True]).reset_index()
```

```python
faixa_por_nivel
```

```python
# plotação do gráfico
sns.set()
g = sns.catplot(
    data=faixa_por_nivel, kind="bar",
    x="nivel", y="quantidade", hue="faixa salarial",
    ci="sd", palette="bright", alpha=.6, height=6
)

g.despine(left=True)
g.set_axis_labels("Nivel", "Quantidade")
g.legend.set_title("Distribuição de Salarios por Nivel")
sns.set(font_scale = 1)
```
